# 서울시 주거지역 화재취약 구역 선정

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/세미프로젝트/data/구별_합본.csv', encoding='euc-kr')
data.set_index('Unnamed: 0', inplace = True)
data
```

```python
import numpy as np
columns = data.columns
columns

for i in columns:
  if i in ['비상소화장치 설치개수']:
    data[f'{i} 점수'] = data[i].rank(ascending=False).astype(int)
  else:
    data[f'{i} 점수'] = data[i].rank(ascending=True).astype(int)

data
```

```python
data['전체 점수'] = data.filter(like='점수').sum(axis=1)
data.head(1)
```

```python
data = data.sort_values(by='전체 점수', ascending = False)
data
```

```python
data.to_csv('include_total_rank.csv')
```

```python
import pandas as pd
import folium


data = pd.read_csv('/content/drive/MyDrive/세미프로젝트/data/include_total_rank.csv', encoding='utf-8')
geo_path = '/content/drive/MyDrive/세미프로젝트/data/서울_자치구_경계_2017.geojson'

# 서울시 중심부의 위도와 경도로 지도 초기화
seoul_map = folium.Map(location=[37.5642135, 127.0016985], zoom_state = 11)

#Choropleth 레이어 추가
folium.Choropleth(
    geo_data=geo_path,
    name='choropleth',
    data=data,
    columns=['Unnamed: 0', '전체 점수'],
    key_on='feature.properties.SIG_KOR_NM',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    legend_name='서울시 취약 분야별 점수 합계(높은 값 일수록 취약)',
    bins=25,
    show_legend=False
).add_to(seoul_map)

# 지도 저장(HTML 파일로 저장하고자 할 경우 경로를 지정)
# seoul_map.save('/mnt/data/seoul_fire_extinguishers_ranking_map.html')

# 지도 직접 출력
# seoul_map

```
